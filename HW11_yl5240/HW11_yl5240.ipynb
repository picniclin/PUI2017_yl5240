## 1.3 Merge data from 1994-2014

```python
year = list(np.arange(94,100)) + list(np.arange(0,15))
year = [str(i).zfill(2) for i in year]
year
```

```python
for i in year:
    fname = 'zbp' + i + 'totals.zip'
    zf = zipfile.ZipFile(fname)
    df = pd.read_csv(zf.open(fname.replace('.zip','.txt')))
    print('year {} columns:{}'.format(i, list(df.columns)))
```

## Because there is no variables of 'stabbr' and  'cty_name' to be used to select all the zipcodes in NYC, like I did in the last part, and the variable 'name' has no enough information to identify the county especially the Queens county, so I had only select the zipcodes in the NYC shapefile(including the three overlapped with Nassau county). The zipcodes not in NYC shapefile cannot be included.

```python
nyc_est = pd.DataFrame()
for i in year:
    fname = 'zbp' + i + 'totals.zip'
    zf = zipfile.ZipFile(fname)
    df = pd.read_csv(zf.open(fname.replace('.zip','.txt')))
    if df.columns[0] == 'ZIP':
        df.rename(columns = {'ZIP':'zip','EST':'est'}, inplace = True)
    temp = df[df.zip.isin(ziplist)][['zip','est']]
    temp.columns = ['zip', 'est_'+ i]
    print('In the year {}, there are {} zipcodes belonged to NYC and in NYC shapefiles.'.format(i,len(temp)))
    if len(nyc_est) == 0: #the first time
        nyc_est = temp
    else:
        nyc_est = nyc_est.merge(temp, on = 'zip', how = 'outer')
    print('{} accumulatived zipcodes.'.format(len(nyc_est)))   
```

## 1.4 Standardize

```python
def standardize(x):
    return ((x - x.mean())/x.std())
```

```python
nyc_est_std = standardize(nyc_est).iloc[:,1:-1]
```

```python
nyc_est_std.head()
```

```python
nyc_est_std.columns = np.arange(1994,2015)
nyc_est_std.head()
```

## the whole picture of normed est, 21 years time series trend for 245 zipcodes

```python
plt.figure(figsize=(12,8))
plt.plot(nyc_est_std.T)
plt.xticks((1994,1998,2002,2006,2010,2014),('1994','1998','2002','2006','2010','2014'))
plt.show()
```
